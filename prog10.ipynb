```
df1 = pd.DataFrame({'Item':['A','B','C','D'],
                    'Value': [1,2,3,4]})
df2 = pd.DataFrame({'Item':['E','F','G','H'],
                    'Quantity': [5,6,7,8]})
pd.concat([df1,df2]).reset_index()
```

```
df1 = pd.DataFrame({'Item':['A','B','C','D'],
                    'Value': [1,2,3,4]})
df2 = pd.DataFrame({'Item':['E','F','G','H'],
                    'Quantity': [5,6,7,8]})
pd.concat([df1,df2], axis=1)
```

using append to add multiple dataframes to one another

```
df1 = pd.DataFrame({'Item':['A','B','C','D'],
                    'Value': [1,2,3,4]})
df2 = pd.DataFrame({'Item':['E','F','G','H'],
                    'Quantity': [5,6,7,8]})
df3 = pd.DataFrame({'Item':['I','J','K','L'],
                    'Quantity': [9,10,11,12]})
df1.append([df2,df3])
```

```
dfA = pd.DataFrame({'key':['A','B','C','D'],
                    'A': [1,2,3,4]})
dfB = pd.DataFrame({'key':['E','F','G','H'],
                    'B': [5,6,7,8]})
pd.merge(dfA,dfB, on='key')
```

```
dfA = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A3','A4'],
                    'B': ['B0','B1','B3','B4']})
dfB = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C3','C4'],
                    'D': ['D0','D1','D3','D4']})
pd.merge(dfA,dfB, on='key')
```

```
dfA = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A3','A4'],
                    'B': ['B0','B1','B3','B4']})
dfB = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C3','C4'],
                    'D': ['D0','D1','D3','D4']})
pd.merge(dfA,dfB, how='left', on='key')
```

```
dfA = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A3','A4'],
                    'B': ['B0','B1','B3','B4']})
dfB = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C3','C4'],
                    'D': ['D0','D1','D3','D4']})
pd.merge(dfA,dfB, how='right', on='key')
```

```
dfA = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A3','A4'],
                    'B': ['B0','B1','B3','B4']})
dfB = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C3','C4'],
                    'D': ['D0','D1','D3','D4']})
pd.merge(dfA,dfB, how='left', on='key')
```

```
dfA = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A3','A4'],
                    'B': ['B0','B1','B3','B4']})
dfB = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C3','C4'],
                    'D': ['D0','D1','D3','D4']})
pd.merge(dfA,dfB, how='inner', on='key')
```

```
22# Program for finding LCM & HCF of two input numbers
def computehcf(x,y):
  #choose the smaller number
    if x>y:
      smaller=y
    else:
      smaller=y
    for i in range(1,smaller+1):
      if((x%i==0) and (y%i==0)):
        hcf=i
    return hcf      

num1=int(input("enter the first number\t"))
num2=int(input("enter the second number\t"))
print("the hcf of",num1,"and",num2,"is",computehcf(num1,num2))
```

```
# python program to find the lcm and hcf of two input numbers
def lcm(x,y):

    if x>y:
        greater = x
    else:
        greater = y
    while(True):
        if ((greater % x == 0)and (greater % y == 0)):
            lcm = greater
            break
        greater += 1
    return lcm
num1 = int(input("enter first number"))
num2 = int(input("enter second number"))
print("The LCM of", num1, "and",num2, "is", lcm(num1,num2) )
```
