**Conditional Statements-IF and ELSE**

```
#using the if statement
x=5
if x<10:
  print("less then 10")
else:
  print("over 10")
```

```
#using the if statement 
x=1
if x<10:
  print("less than 10")
elif x>10 and x<=19:
    print("greater than ten and less than twenty")
else:
  print("over 20")
```
