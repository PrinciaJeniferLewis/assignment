```
x=0
while x<5:
  x+=1
  print(x)
```

Functions

```
#simple functions to return the area of the circle
def AreaCircle(radius):
  area=3.14*(radius **2)
  return area
```

```
AreaCircle(10)
```

```
#simple Functions to return volume of cylinder
def VolumeCylinder(radius,height):
  volume=3.14*(radius **2)*height
  return volume
```

```
VolumeCylinder(10,50)
```
